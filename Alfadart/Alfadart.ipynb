```python
def data_list(data):
    data_start=dt.datetime(2021, 11, 1)
    data_end=dt.datetime(2021, 11, 15)
    t = (data_end - data_start)
    n = int(t.days) + 1
    print(n)
    data['Requests'] = data['Requests'] / n
    data['Impressions'] = data['Impressions'] / n
    data['Revenue'] = data['Revenue'] / n
    data['Дата'] = data_start
    d = {'ID': [], 'AdUnit': [], 'Requests': [], 'Impressions': [], 'Revenue': [], 'size': [], 'Дата': []}
    new_data= pd.DataFrame(data=d)
    while n > 0:
        data['Дата'] = data_start
        data_start = data_start + dt.timedelta(days=1)
        new_data = new_data.append(data.copy())
        n = n - 1
    return new_data
```
